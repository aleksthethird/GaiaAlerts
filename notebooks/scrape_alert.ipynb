```python
def plot_one_spectrum(data, alertname, lims, fn):
    fig, ax = plt.subplots(2, 1, figsize=(6, 6))
    ax[0].set_title(alertname)
    ax[0].plot(data['bp'], color="b", drawstyle="steps-mid")
    ax[0].set_ylabel("Bp flux (Gaia units)")
    ax[0].set_ylim(-0.1 * lims[0], lims[0])
    ax[0].axhline(0., color="k", alpha=0.5, zorder = -10.)
    ax[1].plot(data['rp'], color="r", drawstyle="steps-mid")
    ax[1].set_ylabel("Rp flux (Gaia units)")
    ax[1].set_xlabel("spectral pixel number (uncalibrated)")
    ax[1].set_ylim(-0.1 * lims[1], lims[1])
    ax[1].axhline(0., color="k", alpha=0.5, zorder = -10.)
    fig.savefig(fn)
    plt.close(fig)
    return

def plot_all_spectra(data, alertname):
    ndata = len(data)
    maxes = np.zeros((ndata, 2))
    for i, dd in enumerate(data):
        maxes[i, 0] = np.max(dd['bp'])
        maxes[i, 1] = np.max(dd['rp'])
    lims = np.percentile(maxes, 90., axis=0) # so embarassed !!
    for i, dd in enumerate(data):
        fn = "{}_{:03d}".format(alertname, i)
        plot_one_spectrum(dd, alertname, lims, fn)
    return
```

```python
plot_all_spectra(data, alertname)
```
